```python
def get_ax_size(ax):
    bbox = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    width, height = bbox.width, bbox.height
    width *= fig.dpi
    height *= fig.dpi
    return width, height

def make_image_heatmap_multipanel(expression_df_in,
                       diff_exp_df_in,
                       genes_panels_lst, #list of lists like this [[genes],[genes],...]
                       sample_group_series,
                       title = "Title",
                       small_title_text = None,
                       score_column_name="Score",
                       display_score_column_name = "IC",
                       p_value_column_name = "P-Value",
                       minimum_pval_limit = 0.001,
                       expression_colormap="bwr",
                       output_path = None,
                       group_on_right = None,
                       show_colorbar = False,
                       IC_column_x_coord = None,
                       p_value_column_x_coord = None,
                       fig_size = (25,20),
                       group_key_vs_bottom_height_ratio = None,
                       graph_min_z_score=None,
                       graph_max_z_score=None,
                       dpi = 150,
                       square = False,
                       title_font_size = 30,
                       small_title_font_size = 18,
                       info_font_size = 18,
                       feature_font_size = 18,
                       header_font_size = 18):
    
    if group_key_vs_bottom_height_ratio is None:
        group_key_vs_bottom_height_ratio = [1] + [len(k) for k in genes_panels_lst]
    sup_title_position = 0
    for i in group_key_vs_bottom_height_ratio:
        sup_title_position = sup_title_position + i
    fig, all_axes_lst = plt.subplots(1 + len(genes_panels_lst),1,figsize=fig_size,gridspec_kw={'height_ratios': group_key_vs_bottom_height_ratio,'width_ratios':[1]})
    group_label_right = list(set(list(sample_group_series)))[1]
    if group_on_right is not None:
        group_label_right = group_on_right
    sample_group_series_ordered = sample_group_series.apply(lambda x:binarize_group(x,group_label = group_label_right)).sort_values(ascending = True)
    print(group_key_vs_bottom_height_ratio)
    sns.heatmap(pd.DataFrame(sample_group_series_ordered).T,cmap='Greys',center=0.25,cbar=False,ax=all_axes_lst[0],square=square)
    
    all_axes_lst[0].xaxis.set_ticks([])
    all_axes_lst[0].set_xlabel('')
    for i in range(1,len(all_axes_lst)):
        all_axes_lst[i].set_xlabel('')
    all_axes_lst[0].yaxis.set_ticks([])

    n_samples = len(sample_group_series_ordered.index)
    plt.suptitle(title,fontsize=title_font_size,y=1.23)
    if small_title_text is not None:
        plt.title(small_title_text,fontsize=small_title_font_size,y = 1.055, pad=60)
    
    
    
    if IC_column_x_coord is None:
        IC_column_x_coord = n_samples + (0.15/4)*n_samples + 0.1 #0.3
    if p_value_column_x_coord is None:
        p_value_column_x_coord = n_samples + (0.45/4)*n_samples + 0.3 #0.3

    if square==True:
        if IC_column_x_coord is None:
            IC_column_x_coord = n_samples + (0.5/4)*n_samples + 0.1 #0.3
        if p_value_column_x_coord is None:
            p_value_column_x_coord = n_samples + (0.8)*n_samples + 0.3 #0.3

    all_axes_lst[1].text(IC_column_x_coord,-0.4,display_score_column_name,fontsize=header_font_size,verticalalignment='center',horizontalalignment='center',fontweight='bold')
    all_axes_lst[1].text(p_value_column_x_coord,-0.4,"P-Value",fontsize=header_font_size,verticalalignment='center',horizontalalignment='center',fontweight='bold')
    
    ax_counter = 1
    for genes_panel_lst in genes_panels_lst:
        genes_to_plot = diff_exp_df_in[score_column_name].loc[[g for g in genes_panel_lst if g in diff_exp_df_in.index]].dropna().sort_values(ascending=False).index
        expression_df = expression_df_in.loc[genes_to_plot]
        diff_exp_df = diff_exp_df_in.loc[genes_to_plot]
        
        sns.heatmap(z_normalize_group(expression_df.loc[genes_to_plot][sample_group_series_ordered.index]),
                    cmap =expression_colormap,
                    center=0,cbar=show_colorbar,
                    ax=all_axes_lst[ax_counter],
                    square=square,
                    #cbar_kws = dict(use_gridspec=True,location="top"),
                    vmin=graph_min_z_score,vmax=graph_max_z_score)
        
        print("Axis {}: {} {} | {}".format(ax_counter,all_axes_lst[ax_counter].bbox.height, all_axes_lst[ax_counter].bbox.width,all_axes_lst[ax_counter].bbox.width/all_axes_lst[ax_counter].bbox.height))
        all_axes_lst[ax_counter].set_yticklabels(all_axes_lst[ax_counter].get_yticklabels(),fontsize=feature_font_size, rotation = 0)
        all_axes_lst[ax_counter].xaxis.set_ticks([])

        plt.subplots_adjust(top=1.2,hspace=0.025,wspace=0.25)
        
        y_tick_locations = all_axes_lst[ax_counter].get_yticks()
        

        for i in range(0,len(genes_to_plot)):
            gene_name = genes_to_plot[i]
            gene_info_df = diff_exp_df.loc[gene_name]
            score = round(gene_info_df[score_column_name],3)
            p_value = round(gene_info_df[p_value_column_name],3)
            p_value_txt = str(p_value).ljust(5, "0")
            score_txt = str(score).ljust(5, "0")
            if score < 0:
                score_txt = str(score).rjust(6, "0")
            if p_value < minimum_pval_limit:
                p_value_txt = "< {}".format(minimum_pval_limit)
            all_axes_lst[ax_counter].text(IC_column_x_coord,
                     y_tick_locations[i],
                     score_txt,fontsize=info_font_size,verticalalignment='center',horizontalalignment='center')
            all_axes_lst[ax_counter].text(p_value_column_x_coord,
                     y_tick_locations[i],
                     p_value_txt,fontsize=info_font_size,verticalalignment='center',horizontalalignment='center')
        ax_counter = ax_counter + 1
    '''stats_table = pd.plotting.table(ax_bottom,diff_exp_df.loc[genes_to_plot][["Score","P-Value"]],loc='right',colWidths=[0.1,0.1])
    stats_table.set_fontsize(18)
    stats_table.scale(1, 4)'''
    #plt.xlabel('Mean Decision Time')
    #plt.ylabel('Error Rate')
    if output_path is not None:
        fig.savefig(output_path,dpi=dpi,bbox_inches = 'tight')
    plt.show()
```

```python
import scipy.cluster.hierarchy as shc
def hierarchical_cluster_on_indexes(df_in):
    Z = shc.linkage(df_in.fillna(0), 'ward')
    samples_cluster_ordered_indexes = shc.dendrogram(Z,no_plot = True)["leaves"]
    return df_in.iloc[samples_cluster_ordered_indexes]

def z_normalize_group(exp_df_in,do_clip = True,do_shift = False,do_rank = False):
    exp_df_in_norm = exp_df_in.copy()
    exp_df_in_norm = exp_df_in_norm[~exp_df_in_norm.index.duplicated(keep='first')]
    if do_rank==True:
        exp_df_in_norm = exp_df_in_norm.rank(axis=0, method='average', numeric_only=None, na_option='keep', ascending=True, pct=False)
    exp_df_in_means = exp_df_in_norm.mean(axis=1)
    exp_df_in_stds = exp_df_in_norm.std(axis=1)
    for i in exp_df_in_norm.index:
        #print(exp_df_in_norm.loc[i,:])
        exp_df_in_norm.loc[i,:] = (exp_df_in_norm.loc[i,:] - exp_df_in_means.loc[i])/exp_df_in_stds.loc[i]
    if do_clip==True:
        exp_df_in_norm.clip(-2, 2, inplace=True) 
    if do_shift==True:
        exp_df_in_norm = exp_df_in_norm + 2
    return exp_df_in_norm

def make_image_heatmap_multi_group(expression_df_in,
                       diff_exp_df_dict_in,
                       genes_to_plot_in,
                       sample_group_series,
                                   sample_groups_ordered,
                                   comparison_groups_ordered,
                       title = "Title",
                       small_title_text = None,
                       score_column_name="Score",
                       display_score_column_name = "IC",
                       p_value_column_name = "P-Value",
                       minimum_pval_limit = 0.001,
                       expression_colormap="bwr",
                                sample_group_colormap="Greys",
                                   sample_group_colormap_center=0.25,
                       output_path = None,
                       do_clustering = True,
                       show_colorbar = False,
                       IC_column_x_coord_lst = None,
                       p_value_column_x_coord_lst_in = None,
                        distance_between_comparison_result_groups = 2.5,
                       fig_size = (25,20),
                       group_key_vs_bottom_height_ratio = [1, 25],
                       graph_min_z_score=None,
                       graph_max_z_score=None,
                       dpi = 150,
                       square = False,
                       title_font_size = 30,
                       small_title_font_size = 18,
                       info_font_size = 18,
                       feature_font_size = 18,
                       header_font_size = 18):
    
    display_df_lst = []
    for sample_group_name in comparison_groups_ordered:
        diff_exp_df = diff_exp_df_dict_in[sample_group_name]
        display_df_lst.append(diff_exp_df.loc[[g for g in genes_to_plot_in if g in diff_exp_df.index]][[score_column_name, p_value_column_name]])
    #display_df = pd.concat(display_df_lst,axis=1)
    genes_to_plot = [g for g in genes_to_plot_in if g in expression_df_in.index]
    expression_df_unclustered = expression_df_in.loc[genes_to_plot]
    expression_df = expression_df_unclustered.copy()
    if do_clustering:
        expression_df = hierarchical_cluster_on_indexes(expression_df_unclustered)
    #diff_exp_df = diff_exp_df_in.loc[genes_to_plot]
    
    fig, [ax_top,ax_bottom] = plt.subplots(2,1,figsize=fig_size,gridspec_kw={'height_ratios': group_key_vs_bottom_height_ratio})
    
    sample_group_series_ordered = sample_group_series.apply(lambda x:sample_groups_ordered.index(x)).sort_values(ascending = True)
    print(sample_group_series_ordered)
    sns.heatmap(pd.DataFrame(sample_group_series_ordered).T,cmap=sample_group_colormap,center=sample_group_colormap_center,cbar=False,ax=ax_top,square=square)
    
    ax_top.xaxis.set_ticks([])
    ax_top.set_xlabel('')
    ax_bottom.set_xlabel('')
    ax_top.yaxis.set_ticks([])

    n_samples = len(sample_group_series_ordered.index)
    plt.suptitle(title,fontsize=title_font_size)
    if small_title_text is not None:
        plt.title(small_title_text,fontsize=small_title_font_size,y = 1.055, pad=60)
    sns.heatmap(z_normalize_group(expression_df[sample_group_series_ordered.index]),cmap =expression_colormap,center=0,cbar=show_colorbar,ax=ax_bottom,square=square,
                cbar_kws = dict(use_gridspec=True,location="top"),vmin=graph_min_z_score,vmax=graph_max_z_score)
    
    ax_bottom.set_yticklabels(ax_bottom.get_yticklabels(),fontsize=feature_font_size, rotation = 0)
    ax_bottom.xaxis.set_ticks([])
    
    plt.subplots_adjust(top=0.945,hspace=0.025,wspace=0.15)
    y_tick_locations = ax_bottom.get_yticks()
    
    p_value_column_x_coord_lst = []
    if p_value_column_x_coord_lst_in is not None:
        p_value_column_x_coord_lst = [i for i in p_value_column_x_coord_lst_in]
    if square==True:
        if IC_column_x_coord_lst is None:
            IC_column_x_coord_lst = []
            for i in range(1,len(sample_groups_ordered)+1):
                if i==1:
                    IC_column_x_coord = n_samples + (i*1.75)  #0.3
                else:
                    IC_column_x_coord = p_value_column_x_coord_lst[-1] + distance_between_comparison_result_groups
                IC_column_x_coord_lst.append(IC_column_x_coord)
                if p_value_column_x_coord_lst_in is None:
                    p_value_column_x_coord = IC_column_x_coord + 2.5
                    p_value_column_x_coord_lst.append(p_value_column_x_coord)
        #if p_value_column_x_coord_lst is None:
        #    p_value_column_x_coord_lst = [c + 2 for c in IC_column_x_coord_lst] #n_samples + (0.8)*n_samples + 0.3 #0.3
    else:
        if IC_column_x_coord_lst is None:
            IC_column_x_coord_lst = []
            for i in range(1,len(sample_groups_ordered)+1):
                IC_column_x_coord = n_samples + (0.15/4)*(n_samples+i) + 0.1 #0.3
                IC_column_x_coord_lst.append(IC_column_x_coord)
        if p_value_column_x_coord_lst is None:
            #p_value_column_x_coord = n_samples + (0.45/4)*n_samples + 0.3 #0.3
            p_value_column_x_coord_lst = [c + 0.2 for c in IC_column_x_coord_lst]
    print("IC_column_x_coord_lst: {}".format(IC_column_x_coord_lst))
    print("p_value_column_x_coord_lst: {}".format(p_value_column_x_coord_lst))
    #print([c + 2 for c in IC_column_x_coord_lst])
    for comparison_index in range(0,len(comparison_groups_ordered)):
        IC_column_x_coord = IC_column_x_coord_lst[comparison_index]
        p_value_column_x_coord = p_value_column_x_coord_lst[comparison_index]
        plt.text(IC_column_x_coord,-0.4,display_score_column_name,fontsize=header_font_size,verticalalignment='center',horizontalalignment='center',fontweight='bold')
        plt.text(p_value_column_x_coord,-0.4,"P-Value",fontsize=header_font_size,verticalalignment='center',horizontalalignment='center',fontweight='bold')
        comparison_name = comparison_groups_ordered[comparison_index]
        diff_exp_df = display_df_lst[comparison_index] #diff_exp_df_dict_in[comparison_name]
        title_x_coord = (IC_column_x_coord+p_value_column_x_coord)/2
        plt.text(title_x_coord,-0.9,comparison_name,fontsize=header_font_size,verticalalignment='center',horizontalalignment='center',fontweight='bold')
        for i in range(0,len(expression_df.index)):
            gene_name = expression_df.index[i]
            score_txt = "NaN"
            p_value_txt = "NaN"
            if gene_name in diff_exp_df.index:
                gene_info_df = diff_exp_df.loc[gene_name]
                score = round(gene_info_df[score_column_name],3)
                p_value = round(gene_info_df[p_value_column_name],3)
                p_value_txt = str(p_value).ljust(5, "0")
                score_txt = score_txt = str(score).ljust(5, "0")
                if score < 0:
                    score_txt = str(score).ljust(6, "0")
                
                    
                if p_value < minimum_pval_limit:
                    p_value_txt = "< {}".format(minimum_pval_limit)
            
            plt.text(IC_column_x_coord,
                     y_tick_locations[i],
                     score_txt,fontsize=info_font_size,verticalalignment='center',horizontalalignment='center')
            plt.text(p_value_column_x_coord,
                     y_tick_locations[i],
                     p_value_txt,fontsize=info_font_size,verticalalignment='center',horizontalalignment='center')

    '''stats_table = pd.plotting.table(ax_bottom,diff_exp_df.loc[genes_to_plot][["Score","P-Value"]],loc='right',colWidths=[0.1,0.1])
    stats_table.set_fontsize(18)
    stats_table.scale(1, 4)'''
    #plt.xlabel('Mean Decision Time')
    #plt.ylabel('Error Rate')
    if output_path is not None:
        fig.savefig(output_path,dpi=dpi,bbox_inches = 'tight')
    plt.show()
```

```python
from statsmodels.stats.multitest import fdrcorrection
def add_corrected_p_vals(results_df, p_val_column,corrected_pval_column_name):
    genes_w_pvals = results_df[p_val_column].dropna()
    corrected_pvals = pd.Series(fdrcorrection(genes_w_pvals)[1],index=genes_w_pvals.index,name=corrected_pval_column_name)
    return pd.concat([results_df[[c for c in results_df.columns if c !=corrected_pval_column_name]],corrected_pvals],axis=1)
```
